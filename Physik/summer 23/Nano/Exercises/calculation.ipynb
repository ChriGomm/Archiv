```python
import scipy.constants as cons
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dia(x):
    return 0.0783*np.sqrt(x[:,0]**2+x[:,1]**2+x[:,0]*x[:,1])
def ang(x):
    return 180/np.pi*np.arctan(np.sqrt(3)*x[:,1]/(2*x[:,0]+x[:,1]))

c_id = np.array([[5,0],[6,3],[3,1],[2,2]])
d= dia(c_id)
print(150/d+30)
print(dia(c_id))
print(dia(c_id)*np.pi)
print(ang(c_id))
w0 = 2.07
wp_D = 8.95
gamD = 0.0658
wp_B = 2.96
gamB = 0.59
def hbarlam(lam):
    return cons.c*cons.hbar*np.pi*2/lam*10**9/cons.eV
def epsD_real(lam):
    return 1-wp_D**2/(hbarlam(lam)**2+gamD**2)
def epsD_img(lam):
    return wp_D**2*hbarlam(lam)*gamD/(hbarlam(lam)**4+gamD**2*hbarlam(lam)**2)
def epsB_real(lam):
    return 1-wp_B**2*(hbarlam(lam)**2/w0**2)/((hbarlam(lam)**2-w0**2)**2+hbarlam(lam)**2*gamB**2)
def epsB_img(lam):
    return wp_B**2*(hbarlam(lam)*gamB)/((hbarlam(lam)**2-w0**2)**2+hbarlam(lam)**2*gamB**2)

lmbda = np.linspace(400,800)
fig, ax = plt.subplots(1,1)
ax.plot(lmbda,epsD_real(lmbda))
ax.plot(lmbda,epsD_img(lmbda))
ax.plot(lmbda,epsB_real(lmbda))
ax.plot(lmbda,epsB_img(lmbda))
plt.legend([r"$\epsilon_{D,real}$",r"$\epsilon_{D,imaginary}$",r"$\epsilon_{B,real}$",r"$\epsilon_{B,imaginary}$"])
plt.show()
fig, ax = plt.subplots(1,1)
ax.plot(lmbda,epsD_img(lmbda))
plt.legend([r"$\epsilon_{D,imaginary}$"])
plt.show()
```
